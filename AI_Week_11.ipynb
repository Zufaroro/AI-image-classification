```python
import cv2
import matplotlib.pyplot as plt

# Membaca citra
img = cv2.imread('gambar.jpeg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Konversi ke RGB

# Menampilkan citra asli
plt.imshow(img)
plt.title('Citra Asli')
plt.axis('off')
plt.show()
```

```python
def rotate_image(image, angle):
  height, width = image.shape[:2]
  center = (width // 2, height // 2)
  rotation_matrix = cv2.getRotationMatrix2D(center, angle, 1)
  rotated = cv2.warpAffine(image, rotation_matrix, (width, height))
  return rotated

#rotasi 45 derajat
rotated_img = rotate_image(img, 45)

# Menampilkan hasil
plt.imshow(rotated_img)
plt.title('Rotasi 45 Derajat')
plt.axis('off')
plt.show()
```

```python
#flipping horizontal
flipped_horizontally = cv2.flip(img, 1)

#flipping vertikal
fliiped_vertically = cv2.flip(img, 0)

#menampilkan hasil
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(flipped_horizontally)
plt.title('flipped horizontally')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(fliiped_vertically)
plt.title('flipped vertically')
plt.axis('off')

plt.show()
```

```python
def adjust_brightness(image, factor):
  return cv2.convertScaleAbs(image, alpha=factor, beta=0)

#meningkatkan kecerahan
brighter_img = adjust_brightness(img, 1.5)

#menurunkan kecerahan
darker_img = adjust_brightness(img, 0.5)

#menampilkan hasil
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(brighter_img)
plt.title('kecerahan ditingkatkan')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(darker_img)
plt.title('kecerahan dikurangi')
plt.axis('off')
```

```python
import numpy as np

def add_noise(image):
  noise = np.random.normal(0, 255, image.shape).astype(np.uint8)
  noisy_image = cv2.add(image,noise)
  return noisy_image

#menambahkan noise
noisy_image = add_noise(img)

#menampilkan hasil
plt.imshow(noisy_image)
plt.title('image bernoise')
plt.axis('off')
plt.show()
```

```python
def scale_image(image, scale_factor):
  height,width = image.shape[:2]
  new_width = int(width*scale_factor)
  new_height = int(width*scale_factor)
  return cv2.resize(image, (new_width,new_height))

#scalling 50%
scalled_image = scale_image(img, 0.5)

#menampilkan hasil
plt.imshow(scalled_image)
plt.title('image diperbersar 50%')
plt.axis('off')
plt.show()
```

```python
# 1.	Terapkan kombinasi dari beberapa augmentasi (misalnya, rotasi + flipping).
# 2.	Gunakan citra yang berbeda dan analisis bagaimana augmentasi memengaruhi tampilan citra.
# 3.	Simpan hasil augmentasi dalam format gambar baru.

from PIL import Image

def batman_maker(image):
  height, width = image.shape[:2]
  center = (width // 2, height // 2)
  rotation_matrix = cv2.getRotationMatrix2D(center, 180, 1)
  rotated = cv2.warpAffine(image, rotation_matrix, (width, height))
  darker = adjust_brightness(rotated, 0.5)
  noise = np.random.normal(0, 25, image.shape).astype(np.uint8)
  noisy_image = cv2.add(darker, noise)
  return noisy_image

#menggunakan fungsi
batman = batman_maker(img)

#menampilkan hasil
plt.imshow(batman)
plt.title('image di rotasi dan diberi noise')
plt.axis('off')
plt.show()

# Membuka gambar
image = Image.open("gambar.jpeg")  # Nama file gambar awal

# Menyimpan gambar dalam format baru
image.save("gambar.png", format="PNG")  # Format baru, misalnya PNG

print("Konversi dari jpeg ke png")
```
